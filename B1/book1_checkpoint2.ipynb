```python
from selenium import webdriver
import pandas as pd
import numpy as np
from selenium.webdriver.firefox.options import Options
import time

options = Options()
options.headless = True
```

```python
import os
cwd = os.getcwd()
print(cwd)
da = pd.read_csv('../links.csv')
df = da
url_list = df['url'].tolist() 
print(url_list[147])
```

```python

#url = df.loc[0,0] # df.loc[row_indexer,column_indexer]

# first dictionary 1....1000: 1(1...12), 2(1...12),...
# element_list = []
# details_list = []
myDict1 = {}
start_time = time.time()
  
for page_num in range(0, 2):
    print(page_num,end = " ")
    browser = webdriver.Firefox(options=options, executable_path="./drivers/geckodriver")
    browser.get(url_list[page_num])
    browser.find_element_by_partial_link_text("More Details...").click()
    title = browser.find_element_by_id('bookTitle').text
    authorname = browser.find_element_by_class_name("authorName").text 
    avg_rating = browser.find_element_by_xpath("/html/body/div[2]/div[3]/div[1]/div[2]/div[4]/div[1]/div[2]/div[2]/span[2]").text
    num_rating = browser.find_element_by_xpath("/html/body/div[2]/div[3]/div[1]/div[2]/div[4]/div[1]/div[2]/div[2]/a[2]").text
    num_review = browser.find_element_by_xpath("/html/body/div[2]/div[3]/div[1]/div[2]/div[4]/div[1]/div[2]/div[2]/a[3]").text
    details = browser.find_element_by_id("details")
    temp1 = details.text.split("\n")
    
    page = 0
    series = 0
    
    # num rating
    try:
        num_rat = num_rating.split()[0]
    except:
        num_rat = 0
    
    # num review
    try:
        num_rev = num_review.split()[0]
    except:
        num_rev = 0

    places = np.nan
    awards = np.nan

    # total_pages
    try:
        total_pages = [int(s) for s in temp1[0].split() if s.isdigit()]
        total_pages = str(total_pages[0])
    except IndexError:
        total_pages = np.nan

    # year
    try:
        year = [int(s) for s in temp1[1].split() if s.isdigit()]
        ori_year = str(year[0]) 
    except IndexError:
        ori_year = np.nan

    # places
    for i in range(len(temp1)):
        if(temp1[i]=="Setting"):
            places = temp1[i+1]
        
    # series
    for i in range(len(temp1)):
        if(temp1[i] == "Series"):
            series = 1

    # Awards
    for i in range(len(temp1)):
        if(temp1[i] == "Literary Awards"):
            awards = temp1[i+1]
    
    # genre
    genre = browser.find_elements_by_css_selector("div.left>a.bookPageGenreLink")
    temp2 = [i.text for i in genre]

    try:
        genre3 = temp2[0:3]
    except IndexError:
        genre3 = np.nan

    myDict1[page_num] = (url_list[page_num],title, authorname, avg_rating, num_rat,  num_rev,  total_pages, ori_year, places, series, awards, genre3)
    browser.close()

print("time taken: %s seconds ---" % (time.time() - start_time))
df = pd.DataFrame.from_dict(myDict1, orient='index', columns=['url','title', 'authorname', 'avg_rating', 'num_rat', 'num_rev',  'total_pages', 'ori_year', 'places', 'series', 'awards', 'genre3'])
df.to_csv('filename.csv')
```

```python
# importing necessary packages
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
import seaborn as sns

from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by  import By as selenium_By
from selenium.webdriver.support.ui import Select as selenium_Select
from selenium.webdriver.support.ui import WebDriverWait as selenium_WebDriverWait
from selenium.webdriver.support    import expected_conditions as selenium_ec
from selenium.webdriver.firefox.options import Options as selenium_options
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities as selenium_DesiredCapabilities
from selenium import webdriver as selenium_webdriver
import time
from datetime import datetime


options = Options()
options.headless = True
driver = webdriver.Firefox(options=options, executable_path=r'./drivers/geckodriver')
# for holding the resultant list
element_list = []
myDict = {}
  
# Adding list as value
myDict["key1"] = [1, 2]
myDict["key2"] = ["Geeks", "For", "Geeks"] 
  
#print(myDict)
myDict1 = {}
start_time = time.time()
for page in range(1, 3, 1):
    
    page_url = "https://webscraper.io/test-sites/e-commerce/static/computers/laptops?page=" + str(page)
    #driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get(page_url)
    title = driver.find_elements_by_class_name("title")
    price = driver.find_elements_by_class_name("price")
    description = driver.find_elements_by_class_name("description")
    rating = driver.find_elements_by_class_name("ratings")
  
    for i in range(len(title)):
        myDict1[page] = [title[i].text, price[i].text, description[i].text, rating[i].text]
        #element_list.append([title[i].text, price[i].text, description[i].text, rating[i].text])

print(myDict1)
  
#closing the driver
driver.close()

print("time taken: %s seconds ---" % (time.time() - start_time))
```
