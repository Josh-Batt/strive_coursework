```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Corners

```python
image = cv2.imread('img/chess.jpg')
plt.imshow(image)
```

```python
# We will work with the image above
# Convert it to grayscale
grey_scale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect the corners using the cornerHarris
# Finds the corners of a photo
temp = cv2.cornerHarris(grey_scale, 1, 1, 1)

# Display the image with the detected corners 
plt.imshow(temp)
```
