```python
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv(url)
data
```

**Exercise 1: Read the Total profit of all months and show it using a line plot.**

Total profit data provided for each month. The enerated line plot must include the following properties:

X label name = Month Number

Y label name = Total profit

```python
plt.plot(data['month_number'], data['total_profit'], label ="Company profit per month")
plt.xlabel("Month Number")
plt.ylabel("Total Profit")
plt.title("Company profit per month")
```
