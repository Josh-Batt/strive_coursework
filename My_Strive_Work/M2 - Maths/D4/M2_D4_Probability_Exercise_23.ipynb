#### Exercise 23: How many different seven-letter arrangements of the letters in the word HEXAGON can be made if each letter is used only once?

```python
from itertools import permutations
```

```python
word = "HEXAGON"

all_combos = ["".join(x) for x in permutations(word)]

print("This is the number of letter arrangements: ", len(all_combos))
all_combos
```
